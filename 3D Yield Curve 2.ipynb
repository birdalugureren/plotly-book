```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

from layout import create_layout

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
path_data = 'data/yield.csv'
data = pd.read_csv(path_data,index_col=0,parse_dates=['date'])
data = data.rename(columns = {'Unnamed: 1':'sign'})
```

```python
values = data.drop('sign',axis=1).values

dates = data.index

years = data.drop('sign',axis=1).columns.values
```

```python
colorscale = [[0, 'rgb(103,0,31)'],
              [1, 'rgb(5,48,97)']]
```

```python
def surface_color(x):

    x = x.mask(~((data == 1) | (data == 0)))

    x = x.fillna(axis=1,method='ffill').drop('sign',axis=1).replace(-1,0)

    color_table = x.values
    
    return color_table
```

```python
color_table = surface_color(data)
```

```python
surface=go.Surface(surfacecolor=color_table,
           colorscale=colorscale,
           showscale=False,
           x=years,
           y=dates,
           z=values,
          )
```

```python
layout = create_layout('Russia','Yield')
```

```python
fig = go.Figure(data=surface,layout=layout)
```

```python
fig.show()
```
